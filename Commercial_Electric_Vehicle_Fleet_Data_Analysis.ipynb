```python
#Now creating a dictionary such that it has key as PNID and value as vehicle_id

pnid_to_vehicle = {}

# Iterate through the DataFrame and populate the dictionary
for index, row in vehicle_id_to_pnid_df.iterrows():
    vehicle_id = row['VEHICLE_ID']
    pnids = row['PNID']

    # Iterate through the list of pnids for the current row
    for pnid in pnids:
        pnid_to_vehicle[pnid] = vehicle_id
```

```python
sorted_pnid_to_vehicle = {int(key): value for key, value in pnid_to_vehicle.items()}
sorted_items = sorted(sorted_pnid_to_vehicle.items())
final_pnid_to_vehicle_id = {key: value for key, value in sorted_items}
```

```python
from joblib import Parallel, delayed

# Function to find the closest timestamp within the range



def find_closest_charge_level(row, charge_data):
    vehicle_id = row['VEHICLE_ID']
    timestamp = row['TIMESTAMP']
    time_range_start = timestamp - pd.Timedelta(seconds=300)
    time_range_end = timestamp + pd.Timedelta(seconds=300)

    # Filter charge data for the same vehicle and within the time range
    mask = (charge_data['VEHICLE_ID'] == vehicle_id) & (charge_data['TIMESTAMP'] >= time_range_start) & (charge_data['TIMESTAMP'] <= time_range_end)
    filtered_data = charge_data[mask]

    if not filtered_data.empty:
        # Find the closest timestamp
        filtered_data['time_diff'] = (filtered_data['TIMESTAMP'] - timestamp).abs()
        closest_row = filtered_data.loc[filtered_data['time_diff'].idxmin()]
        return closest_row['Charge Level']
    return np.nan

# Apply the function to each row in parallel
def apply_parallel(df, func, charge_data):
    results = Parallel(n_jobs=-1)(delayed(func)(row, charge_data) for _, row in df.iterrows())
    return results
```

```python
# Apply the function to each row in All_Trigger_Events
All_Trigger_Events['Charge Level'] = apply_parallel(All_Trigger_Events, find_closest_charge_level, All_Charge_Level_Data_CLEANED)
```

```python
#Drop the rows where the Charge Level is NaN
FINAL_DATASET_CLEANED = FINAL_DATASET[FINAL_DATASET['Charge Level'].notna()]
FINAL_DATASET_CLEANED.reset_index(drop=True, inplace=True)
```

```python
#For each vehicle ID calculate the difference between the present and previous charge level, for first reading of any vehicle keep it nan
FINAL_DATASET_CLEANED['Charge Level Difference'] = FINAL_DATASET_CLEANED.groupby('VEHICLE_ID')['Charge Level'].diff()
```

```python
FINAL_DATASET_CLEANED
```

```python
# Initialize an empty list to collect charge events
charge_events = []

# Define thresholds
threshold_off = 1
threshold_on = 3

# Loop through each vehicle's data
for vehicle_id in FINAL_DATASET_CLEANED['VEHICLE_ID'].unique():
    vehicle_data = FINAL_DATASET_CLEANED[FINAL_DATASET_CLEANED['VEHICLE_ID'] == vehicle_id]

    charge_start = None
    charge_start_level = None
    charge_end = None
    charge_end_level = None

    for i in range(1, len(vehicle_data)):
        current_row = vehicle_data.iloc[i]
        previous_row = vehicle_data.iloc[i - 1]

        # Check for ignition status and corresponding threshold
        if previous_row['Event'] == 'ignitionOff' and current_row['Charge Level Difference'] > threshold_off:
            if charge_start is None:
                charge_start = previous_row['TIMESTAMP']
                charge_start_level = previous_row['Charge Level']
            charge_end = current_row['TIMESTAMP']
            charge_end_level = current_row['Charge Level']

        elif previous_row['Event'] == 'ignitionOn' and current_row['Charge Level Difference'] > threshold_on:
            if charge_start is None:
                charge_start = previous_row['TIMESTAMP']
                charge_start_level = previous_row['Charge Level']
            charge_end = current_row['TIMESTAMP']
            charge_end_level = current_row['Charge Level']

        # If charge_end is set, append to charge_events and reset variables
        if charge_end is not None:
            charge_events.append({
                'Vehicle id': vehicle_id,
                'charging_event_start': charge_start,
                'charging_event_end': charge_end,
                'Ev_battery_level_start': charge_start_level,
                'Ev_battery_level_end': charge_end_level,
                'Battery_change': charge_end_level - charge_start_level
            })
            charge_start = None
            charge_end = None
            charge_start_level = None
            charge_end_level = None
```

```python
# Create DataFrame from collected events
CHARGE_EVENTS = pd.DataFrame(charge_events)
```
